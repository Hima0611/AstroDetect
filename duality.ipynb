```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# Train
model.train(
    data="/content/drive/MyDrive/MyDrive/Exercise3/dataset/config.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    project="/content/drive/MyDrive/MyDrive/Exercise3/runs",
    name="train_exp13"
)
```

```python

import os
from ultralytics import YOLO
import gradio as gr
from PIL import Image


best_path = "best.pt"


if not os.path.exists(best_path):
    raise FileNotFoundError(f"âŒ Could not find {best_path}! Make sure you uploaded it.")
else:
    print("âœ… Uploaded model found at:", best_path)


model = YOLO(best_path)


def detect(image):
    results = model.predict(source=image, conf=0.5, save=False)
    # Returns image with bounding boxes
    return results[0].plot()


demo = gr.Interface(
    fn=detect,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="numpy"),
    title="YOLOv8 Object Detection",
    description="Upload an image to detect FireExtinguisher, ToolBox, OxygenTank ğŸš€"
)


demo.launch(debug=True, share=True)
```
